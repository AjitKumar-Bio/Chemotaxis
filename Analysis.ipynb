```python
import random
import matplotlib.pyplot as plt

def random_walk(steps):
    position = 0
    positions = [position]
    for i in range(steps):
        step = random.choice([-1, 1])
        position += step
        positions.append(position)
    return positions

def biased_random_walk(steps, p):
    position = 0
    positions = [position]
    for i in range(steps):
        if random.random() < p:
            step = 1
        else:
            step = -1
        position += step
        positions.append(position)
    return positions

# Simulate a random walk with 1000 steps
rw_positions = random_walk(1000)
# Simulate a biased random walk with 1000 steps and a bias towards positive steps (p=0.6)
brw_positions = biased_random_walk(1000, 0.6)

# Plot the results
plt.plot(rw_positions, label='Random Walk')
plt.plot(brw_positions, label='Biased Random Walk')
plt.legend()
plt.show()
```
